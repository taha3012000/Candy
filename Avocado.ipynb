```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('C:/Users/Taha/Downloads/avocado.csv')
data.head()
```

```python
plt.figure(figsize=(10,10))
plt.title('Average price per type')
plt.xticks(rotation=0)
sns.boxplot(x='type' , y = 'AveragePrice' ,data =data,palette='Set1')
#organic are more expensive
```

```python
data.year.value_counts().sort_index().plot(kind='bar' ,color='magenta' , figsize=(4,4))
plt.xticks(rotation=0)
```

```python
#nortce that we work with cities only not states
inconsistent=['California', 'GreatLakes', 'Midsouth', 'NewYork', 'Northeast', 'SouthCarolina', 'Plains', 'SouthCentral', 'Southeast', 'TotalUS', 'West']
data=data[~data.region.isin(inconsistent)]
```

```python
plt.figure(figsize=(15,15))
plt.title('Average price by city')
sns.barplot(x='AveragePrice' , y='region',data=data ,palette='deep')
#avg price is the most in San Francisco and hartford springfield 

```

```python
plt.figure(figsize=(20,20))
plt.title('Average price per region')
sns.boxplot(x="AveragePrice",y="region",data= data,palette="deep")

# Houston  has the cheapest avocados but cheapest price at one time was recoreded in PhoenixTucson
```

```python
plt.figure(figsize=(15,15))
plt.title('Total volume per region')
sns.barplot(x="Total Volume",y="region",data= data,palette="rocket")
#Los Angeles consumed more than any city
```
