```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_excel('C:/Users/Taha/Downloads/candyhierarchy2017.xlsx')
data.head()
```

```python
#to know information about data and missing values (if it exist)
data.info()
```

```python
#to know sum of missing values
data.isnull().sum()
```

```python
#to drop columns which unuseful
data.drop(['Q12: MEDIA [Yahoo]','Q12: MEDIA [ESPN]','Q12: MEDIA [Daily Dish]','Unnamed: 113','Q11: DAY' , 'Q8: DESPAIR OTHER','Q9: OTHER COMMENTS'],axis=1,inplace=True)
data.head()
```

```python
#set data index by'Internal ID' columns to make drop all rows has a nan values in all colmuns be easy
data.set_index('Internal ID',inplace=True)
```

```python
#drop drop all rows has a nan values in all colmuns
data.dropna(how='all',inplace=True)
data.head()
```

```python
#strip columns name
for col in data.columns:
    data=data.rename(columns=lambda x: x.strip('Q : | 0123456789 ? '))
data.head()
```

```python

data.info()
```

```python
#dtype of columns
data.dtypes
```

```python

data.isnull().sum()
```

```python
#convert dtype if 'AGE' column to be numeric to fill nan by statistical methods
data['AGE'] = data['AGE'].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
#distplot 'AGE' to know if it normal distriburion of skewed and select if we fillna by mean of median
plt.figure(figsize=(10,10))
plt.xlim(0,100)
sns.distplot(data['AGE'])
```

```python
#normal distribution so mean and median are same
#median of columns['AGE','MEDIA [Science]']
dfmedian=data[['AGE','MEDIA [Science]']].median()
```

```python
#fill nan values with median value
data[['AGE','MEDIA [Science]']]=data[['AGE','MEDIA [Science]']].fillna(dfmedian)
data.head()
```

```python
#in object columns we fillna with the most freq
data['DRESS'].value_counts()
```

```python
data['DRESS']=data[['DRESS']].fillna('White and gold')
data.head()
```

```python
data['COUNTRY'].value_counts()
```

```python
data['COUNTRY']=data['COUNTRY'].fillna('USA')
data.head()
```

```python
data['GENDER'].value_counts()
```

```python
data['GENDER']=data[['GENDER']].fillna('Male')
data.head()
```

```python
data['GOING OUT'].value_counts()
```

```python
data['GOING OUT']=data[['GOING OUT']].fillna('No')
data.head()
```

```python
data['STATE, PROVINCE, COUNTY, ETC'].value_counts()
```

```python
data['STATE, PROVINCE, COUNTY, ETC']=data[['STATE, PROVINCE, COUNTY, ETC']].fillna('California')
data.head()
```

```python
data['Anonymous brown globs that come in black and orange wrappers\t(a.k.a. Mary Janes)'].value_counts()
```

```python
data['Anonymous brown globs that come in black and orange wrappers\t(a.k.a. Mary Janes)']=data[['Anonymous brown globs that come in black and orange wrappers\t(a.k.a. Mary Janes)']].fillna('DWSPAIR')
data.head()
```

```python
#another easier way .. use lambda function to get the most freq values and fillna columns with this by for loop
for col in data.columns :
    col=lambda x: x.value_counts()
data.fillna(col)
data.head()

```
